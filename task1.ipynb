```python
import time
import matplotlib.pyplot as plt
```

```python
rec_counter = 0

def fib_recursive(n):
    global rec_counter
    rec_counter += 1
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)
```

```python
def fib_iterative(n):
    iter_count = 0
    if n <= 1:
        return n, iter_count
    a, b = 0, 1
    for i in range(2, n + 1):
        iter_count += 1
        a, b = b, a + b
    return b, iter_count
```

```python
n_values = list(range(1, 40))

time_recursive = []
time_iterative = []
iters_recursive = []
iters_iterative = []
```

```python
for n in n_values:
    rec_counter = 0
    start = time.perf_counter()
    fib_recursive(n)
    end = time.perf_counter()
    time_recursive.append(end - start)
    iters_recursive.append(rec_counter)
```

```python
for n in n_values:
    start = time.perf_counter()
    _, iter_count = fib_iterative(n)
    end = time.perf_counter()
    time_iterative.append(end - start)
    iters_iterative.append(iter_count)
```

```python
plt.figure(figsize=(100, 100))
```

```python
plt.subplot(1, 1, 1)
plt.plot(n_values, time_recursive, 'r-o', label='Рекурсия')
plt.plot(n_values, time_iterative, 'b-o', label='Итерация')
plt.xlabel("n")
plt.ylabel("Время (сек)")
plt.title("Зависимость времени вычисления от n")
plt.legend()
plt.grid(True)
```

```python
plt.subplot(1, 1, 1)
plt.plot(n_values, iters_recursive, 'r-o', label='Рекурсия (вызовы)')
plt.plot(n_values, iters_iterative, 'b-o', label='Итерация (итерации)')
plt.xlabel("n")
plt.ylabel("Количество итераций/вызовов")
plt.title("Зависимость количества итераций/вызовов от n")
plt.legend()
plt.grid(True)
```
