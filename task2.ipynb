```python
import random
import time
import matplotlib.pyplot as plt
```

```python
def gnome_sort(arr):
    ops = 0
    index = 1
    while index < len(arr):
        if index == 0 or arr[index - 1] <= arr[index]:
            ops += 1
            index += 1
        else:
            ops += 1
            arr[index], arr[index - 1] = arr[index - 1], arr[index]
            ops += 1
            index -= 1
    return ops, arr
```

```python
def bubble_sort(arr):
    ops = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            ops += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                ops += 1
    return ops, arr
```

```python
def insertion_sort(arr):
    ops = 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            ops += 1
            arr[j + 1] = arr[j]
            j -= 1
        ops += 1
        arr[j + 1] = key
    return ops, arr
```

```python
sizes = list(range(50, 501, 50))

results = {
    'gnome': {'ops': [], 'time': []},
    'bubble': {'ops': [], 'time': []},
    'insertion': {'ops': [], 'time': []}
}
```

```python
for n in sizes:
    base_array = [random.randint(0, n) for _ in range(n)]
    
    arr_copy = base_array.copy()
    start_time = time.perf_counter()
    ops = gnome_sort(arr_copy)[0]
    elapsed_time = time.perf_counter() - start_time
    results['gnome']['ops'].append(ops)
    results['gnome']['time'].append(elapsed_time)
    
    arr_copy = base_array.copy()
    start_time = time.perf_counter()
    ops = bubble_sort(arr_copy)[0]
    elapsed_time = time.perf_counter() - start_time
    results['bubble']['ops'].append(ops)
    results['bubble']['time'].append(elapsed_time)
    
    arr_copy = base_array.copy()
    start_time = time.perf_counter()
    ops = insertion_sort(arr_copy)[0]
    elapsed_time = time.perf_counter() - start_time
    results['insertion']['ops'].append(ops)
    results['insertion']['time'].append(elapsed_time)
```

```python
plt.figure(figsize=(50, 50))
```

```python
plt.subplot(1, 1, 1)
plt.plot(sizes, results['gnome']['ops'], marker='o', label=' ')
plt.plot(sizes, results['bubble']['ops'], marker='s', label=' ')
plt.plot(sizes, results['insertion']['ops'], marker='^', label=' ')
plt.xlabel('  ')
plt.ylabel(' ')
plt.title('  ')
plt.legend()
plt.grid(True)
```

```python
plt.subplot(1, 1, 1)
plt.plot(sizes, results['gnome']['time'], marker='o', label=' ')
plt.plot(sizes, results['bubble']['time'], marker='s', label=' ')
plt.plot(sizes, results['insertion']['time'], marker='^', label=' ')
plt.xlabel('  ')
plt.ylabel(' (.)')
plt.title('  ')
plt.legend()
plt.grid(True)
```
